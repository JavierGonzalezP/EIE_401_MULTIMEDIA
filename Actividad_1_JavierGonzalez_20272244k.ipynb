## 2. Define en python vectores y funciones
<p align="justify">
Debes generar vectores que representan intervalos de tiempo o de una variable.
Debes ingresar las funciones matemáticas y evaluarlas en los vectores generados.

$$
f(x) =  e^{-x^2} 
$$

$$
f(z) =  ysin(x) 
$$

<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt

# Vectores y variables
x_values = np.linspace(-5, 5, 100)  # Rango de valores para x
z_values = np.linspace(0, 2*np.pi, 100)  # Rango de valores para z
y_value = 2.0

# Funciones matemáticas
def f1(x):
    return np.exp(-x**2)

def f2(y, z):
    return y * np.sin(z)

# Evaluar las funciones en los vectores generados
y1_values = f1(x_values)
y2_values = f2(y_value, z_values)

# Graficas de las funciones
plt.figure(figsize=(10, 5))

# Función f(x) = e^(-x^2)
plt.subplot(1, 2, 1)
plt.plot(x_values, y1_values, label=r'$f(x) = e^{-x^2}$', color="green")
plt.title('Función f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim(-4, 4)
plt.legend()

# Función f(z) = y*sin(z)
plt.subplot(1, 2, 2)
plt.plot(z_values, y2_values, label=r'$f(z) = y \cdot \sin(z)$', color="red")
plt.title('Función f(z)')
plt.xlabel('z')
plt.ylabel('f(z)')
plt.xlim(0, 6)
plt.legend()

plt.tight_layout()
plt.show()

```

### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal senoidal
amplitud = 1.0
frecuencia = 1.0  # en Hz
fase = 0  # en radianes
tiempo_inicio = 0
tiempo_fin = 2
numero_puntos = 1000


# Señal senoidal con el ruido
t = np.linspace(tiempo_inicio, tiempo_fin, numero_puntos)
senoide = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)
ruido_gaussiano = 0.2 * np.random.randn(numero_puntos)  # El factor 0.1 controla la intensidad del ruido
senoide_con_ruido_gaussiano = senoide + ruido_gaussiano

# Grafica de la señal senoidal con ruido gaussiano
plt.figure(figsize=(10, 5))
plt.plot(t, senoide_con_ruido_gaussiano, label='Señal con ruido gaussiano', color='orange')
plt.plot(t, senoide, label='Señal senoidal original', color='blue')
plt.title('Señal senoidal con ruido gaussiano')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.xlim(0, 2)
plt.legend()
plt.grid(True)
plt.show()
```

### 3.2 Ruido Rosa
Investiga en que consiste el ruido rosa, cual es la diferencia del ruido gaussiano y demuéstralo.

El ruido rosa(1/f) es un tipo de ruido aleatorio en el cual la densidad espectral de potencia (PSD) es inversamente proporcional a la frecuencia. Esto significa que en un espectro de frecuencia, la energía del ruido rosa disminuye a medida que aumenta la frecuencia.
La diferencia principal entre el ruido rosa y el ruido gaussiano se encuentra en su comportamiento en el dominio de la frecuencia. Mientras que el ruido gaussiano tiene una PSD constante (es decir, es plano en el dominio de la frecuencia), el ruido rosa tiene una PSD que disminuye con la frecuencia. 
La siguiente figura muestra como la PSD del ruido rosa va decayendo a medida que aumenta la frecuencia, mientras que el ruido gaussiano se mantiene constante.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal del ruido rosa
fs = 44100                                       # frecuencia de muestreo
f = np.linspace(1, 10000, 10000)                 # rango de frecuencias 
psd_rosa = 1 / f                                 # densidad espectral de ruido rosa
psd_rosa_db = 10 * np.log10(psd_rosa)            # convertir la densidad espectral a dB
ruido = np.random.randn(len(f)) * 0.8            # generar ruido aleatorio
psd_rosa_con_ruido = psd_rosa_db + ruido         # sumar el ruido a la densidad espectral de ruido rosa

# Parámetros de la señal del ruido gaussiano
amplitud = 1.0
frecuencia = 1.0            # en Hz
fase = 0                    # en radianes
tiempo_inicio = 1
tiempo_fin = 10000
numero_puntos = 1000
t = np.linspace(tiempo_inicio, tiempo_fin, numero_puntos)
senoide = amplitud * np.sin(2 * np.pi * frecuencia * t + fase)
ruido_gaussiano = 0.8 * np.random.randn(numero_puntos)         # generar ruido gaussiano adicional
senoide_con_ruido_gaussiano = senoide + ruido_gaussiano        # agregar el ruido gaussiano a la señal senoidal existente

# Graficar el ruido rosa vs ruido gaussiano
plt.plot(f, psd_rosa_con_ruido, label='Señal con ruido rosa', color='violet')
plt.plot(t, senoide_con_ruido_gaussiano, label='Señal con ruido gaussiano', color='blue')
plt.xscale('log')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Densidad espectral (dB)')
plt.title('Ruido rosa vs Ruido Gaussiano')
plt.xticks([1, 10, 100, 1000, 10000], ['1 Hz', '10 Hz', '100 Hz', '1 kHz', '10 kHz'])
plt.ylim([-50, 10])
plt.xlim([1, 10000])
plt.legend(loc='lower left')
plt.show()

```
