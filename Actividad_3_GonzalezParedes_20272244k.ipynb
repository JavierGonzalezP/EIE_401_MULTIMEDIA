## 1. Generar una imagen de una función trigronométrica (función bidimensional) 
### **Sinusoidal Gratings**

<p align="justify">
En teoría, cualquier imagen puede ser reconstruida a partir de ondas senoidales. Es bueno entender como podemos generar una imagen que nos meustra una onda senoidal. 


```python
import numpy as np
import matplotlib.pyplot as plt

# Se definen los límites del plano
x_min, x_max = -10, 10
y_min, y_max = -4, 4

# Creación de la malla de puntos en un plano xy
x = np.linspace(x_min, x_max, 100)
y = np.linspace(y_min, y_max, 100)
X, Y = np.meshgrid(x, y)

# Función bidimensional
Z = np.sin(X) * np.cos(Y)

# Se grafica la función bidimensional
plt.figure(figsize=(10, 4))
plt.imshow(Z, extent=[x_min, x_max, y_min, y_max], cmap='inferno')
plt.colorbar(label='f(x, y)')  # Barra de colores
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función trigonométrica f(x, y) = sin(x) * cos(y)')
plt.ylim(y_min, y_max)
plt.grid(False)
plt.show()
```

## 2. Generar señales


### 2. Genera una señal: estas señales las vas a utilizar en los puntos siguientes

Genera la señal $$ sinx+cos2x $$ y graficala.


```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(0, 2*np.pi, 1000)

# Señal sin(x) + cos(2x)
y = np.sin(x) + np.cos(2*x)

# Se grafica la señal
plt.figure(figsize=(10, 4))
plt.plot(x, y, label=r'$\sin(x) + \cos(2x)$',color='orange')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la señal $sin(x) + cos(2x)$')
plt.legend()
plt.grid(True)
plt.show()
```

### 2.2 Genera otra señal no pura como la anterior, sino con ruido gaussiano añadido.

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(0, 2*np.pi, 1000)

# Señal sin(x) + cos(2x)
y = np.sin(x) + np.cos(2*x)

# Se genera ruido gaussiano y se agrega a la señal
mu, sigma = 0, 0.1              # Media y desviación estándar para generar ruido 
ruido = np.random.normal(mu, sigma, len(x))
y_con_ruido = y + ruido

# Se grafica la señal con ruido y la señal original
plt.figure(figsize=(10, 4))
plt.plot(x, y_con_ruido, label='Señal con ruido',color='blue')
plt.plot(x, y, label='Señal original', color='orange')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de la señal $sin(x) + cos(2x)$ con ruido gaussiano')
plt.legend()
plt.grid(True)
plt.show()
```

## 3. Análisis en el dominio de Fourier
<p align="justify">
A partir de las señales anteriores, aplica la transformada de Fourier.    


### 3.1 Obtenga la transformada de Fourier para las señales que generaste en el  punto anterior


```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(0, 2*np.pi, 1000)

# Señal sin(x) + cos(2x)
y = np.sin(x) + np.cos(2*x)

# Generar ruido gaussiano y se agrega a la señal
mu, sigma = 0, 0.1              # Media y desviación estándar para generar ruido 
ruido = np.random.normal(mu, sigma, len(x))
y_con_ruido = y + ruido

# Se aplica la transformada de Fourier
fft_y = np.fft.fft(y)
fft_y_con_ruido = np.fft.fft(y_con_ruido)

# Se calculaa las frecuencias correspondientes
frecuencias = np.fft.fftfreq(len(x), x[1] - x[0])

# Se crean subplots para ambas gráficas
fig, axs = plt.subplots(2, 1, figsize=(8.5, 8))

# Grafica del espectro de frecuencia de la señal original
axs[0].plot(frecuencias, np.abs(fft_y), label='Señal original', color='orange')
axs[0].set_xlabel('Frecuencia')
axs[0].set_ylabel('Amplitud')
axs[0].set_title('Espectro de frecuencia de la señal original')
axs[0].legend()
axs[0].grid(True)

# Grafica del espectro de frecuencia de la señal con ruido
axs[1].plot(frecuencias, np.abs(fft_y_con_ruido), label='Señal con ruido', color='blue')
axs[1].set_xlabel('Frecuencia')
axs[1].set_ylabel('Amplitud')
axs[1].set_title('Espectro de frecuencia de la señal con ruido')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

### 3.2 Obten un resultado Normalizado: 
Investiga de que se trata un resultado normalizado y encargate de normalizar el resultado del punto 3.1

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(0, 2*np.pi, 1000)

# Señal sin(x) + cos(2x)
y = np.sin(x) + np.cos(2*x)

# Generar ruido gaussiano y se agrega a la señal
mu, sigma = 0, 0.1              # Media y desviación estándar para generar ruido 
ruido = np.random.normal(mu, sigma, len(x))
y_con_ruido = y + ruido

# Se aplica la transformada de Fourier
fft_y = np.fft.fft(y)
fft_y_con_ruido = np.fft.fft(y_con_ruido)

# Se calculan las frecuencias correspondientes
frecuencias = np.fft.fftfreq(len(x), x[1] - x[0])

# Se normalizan los resultados de la transformada de Fourier
fft_y_normalizado = fft_y / len(x)
fft_y_con_ruido_normalizado = fft_y_con_ruido / len(x)

# Se crean subplots para ambas gráficas
fig, axs = plt.subplots(2, 1, figsize=(8.5, 8))

# Grafica del espectro de frecuencia de la señal original normalizado
axs[0].plot(frecuencias, np.abs(fft_y_normalizado), label='Señal original', color='orange')
axs[0].set_xlabel('Frecuencia')
axs[0].set_ylabel('Amplitud Normalizada')
axs[0].set_title('Espectro de frecuencia normalizado de la señal original')
axs[0].legend()
axs[0].grid(True)

# Grafica del espectro de frecuencia de la señal con ruido normalizado
axs[1].plot(frecuencias, np.abs(fft_y_con_ruido_normalizado), label='Señal con ruido', color='blue')
axs[1].set_xlabel('Frecuencia')
axs[1].set_ylabel('Amplitud Normalizada')
axs[1].set_title('Espectro de frecuencia normalizado de la señal con ruido')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

### 3.3 Encuentra la transformada inversa de Fourier
_solo para una señal de las dos que generaste inicialmente_ En este caso puede usar una librería. Explica en detalle que hace la librería y el resultado que arroja.

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores para x
x = np.linspace(0, 2*np.pi, 1000)

# Señal sin(x) + cos(2x)
y = np.sin(x) + np.cos(2*x)

# Se aplica la transformada de Fourier
fft_y = np.fft.fft(y)

# # Se aplica la transformada inversa de Fourier
ifft_y = np.fft.ifft(fft_y)

# Se crean subplots para ambas gráficas
fig, axs = plt.subplots(2, 1, figsize=(8.5, 8))

# Grafica de la señal original
axs[0].plot(x, y, label='Señal original', color='orange')
axs[0].set_xlabel('x')
axs[0].set_ylabel('f(x)')
axs[0].set_title('Señal original en el dominio del tiempo')
axs[0].legend()
axs[0].grid(True)

# Grafica de la señal recuperada
axs[1].plot(x, ifft_y.real, label='Señal recuperada', color='blue')
axs[1].set_xlabel('x')
axs[1].set_ylabel('f(x)')
axs[1].set_title('Señal recuperada en el dominio del tiempo')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

### 3.4 Escriba un algoritmo para la Transformada Discreta de Fourier(DFT) 2-D y muéstrelo 
_solo para una señal de las dos que generaste inicialmente_
Aquí debe implementar el algoritmo de DFT, es decir, NO puedes usar paquetes externos para calcular la transformada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Se define la señal de entrada (señal inicial)
x = np.linspace(0, 2*np.pi, 1000)
señal = np.sin(x) + np.cos(2*x)

# Se define la matriz de transformación
def matriz_dft(N):
    n = np.arange(N)
    k = n.reshape((N, 1))
    M = np.exp(-2j * np.pi * k * n / N)
    return M

# Se calcula la DFT en 2-D
def dft_2d(señal):
    N = len(señal)
    W = matriz_dft(N)
    return np.dot(W, señal)

# Se aplica la DFT en 2D a la señal
dft_señal = dft_2d(señal)

# Grafica de la magnitud de la DFT en 2D
plt.figure(figsize=(10, 4))
plt.plot(np.abs(dft_señal), color='red')
plt.xlabel('Frecuencia')
plt.ylabel('Magnitud')
plt.title('Transformada Discreta de Fourier (DFT) en 2D')
plt.grid(True)
plt.show()
```

En la señal original se espera ver picos en el espectro de frecuencia correspondientes a las frecuencias de las componentes sinusoidales presentes en la señal y en el caso de la señal con ruido, además de los picos de las frecuencias de las componentes sinusoidales, también hay un aumento en la magnitud de las frecuencias de ruido introducidas.
